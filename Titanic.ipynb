```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.animation as animation
```

```python
def load_and_clean(filepath):
    df=pd.read_csv(filepath)
    df['Age'] = df['Age'].fillna(df['Age'].median())
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
    return df
```

```python
def doingprediction(df):
    prediction=df.drop(columns=['PassengerId','Ticket','Cabin'])
    label=LabelEncoder()
    prediction['Gender'] = label.fit_transform(prediction['Sex'])
    prediction['Embarked'] = label.fit_transform(prediction['Embarked'])
    return prediction
```

```python
def scaling(prediction):
    scaler=StandardScaler()
    scales=scaler.fit_transform(prediction[['Pclass','Age','SibSp','Parch','Fare']])
    X=pd.DataFrame(scales,columns = ['Pclass','Age','SibSp','Parch','Fare'])
    X['Gender']=prediction['Gender']
    X['Embarked']=prediction['Embarked']
    return X
```

```python
def regression(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42,stratify=y)
    model=LogisticRegression(max_iter=1000)
    model.fit(X_train,y_train)
    return model
```

```python
def makes(model,X):
    y_probablity=model.predict_proba(X)[:, 1]
    predict={'Predicted Probablity':y_probablity,'Predicted Survival Logistic':model.predict(X)}
    return predict
```

```python
def results(df,predict):
    data_prediction=doingprediction(df)
    for key, values in predict.items():
        data_prediction[key]=values
    prediction_details = pd.concat([df[['PassengerId','Name','Ticket','Sex','Age','SibSp','Parch']],
                                   data_prediction[['Predicted Survival Logistic','Predicted Probablity']]],
                            axis=1)
    prediction_details.rename(columns={'Sex': 'Gender'},inplace=True)
    survived_log=prediction_details[prediction_details['Predicted Survival Logistic']==1]
    not_survived_log=prediction_details[prediction_details['Predicted Survival Logistic']==0]
    print("\nPassengers predicted to survive = ")
    print(survived_log[['Ticket','Name','Gender','Age','SibSp','Parch']])

    print("\nPassengers predicted not to survive = ")
    print(not_survived_log[['Ticket','Name','Gender','Age','SibSp','Parch']])
    return prediction_details
    
```

```python
def histogram(prediction_details):
    plt.figure(figsize=(15,5))
    ax=plt.gca()
    ax.set_facecolor('Black')
    ax.patch.set_alpha(0.2)

    plt.hist(prediction_details['Predicted Probablity'],bins=30,color='blue',edgecolor='red',alpha=0.8)
    plt.title('Histogram of predicted Probablity from Logistic Regression',fontsize=16,fontweight='bold')
    plt.xlabel('Predicted Probablity',fontsize=14,fontweight='bold')
    plt.ylabel('Frequency',fontsize=14,fontweight='bold')
    plt.grid(True,linestyle='--',alpha=0.7)



    def animate(i):
        edge_colors=['orange','red','yellow']
        plt.gca().patch.set_edgecolor(edge_color[i % len(edge_colors)])
        return plt.gca().patch()
    histogram=animation.FuncAnimation(plt.gcf(),animate,frames=30,interval=100)
    plt.tight_layout()
    plt.show()


```

```python
filepath='Titanic-Dataset.csv'
df=load_and_clean(filepath)
prediction=doingprediction(df)
X=scaling(prediction)
y=df['Survived']
model=regression(X,y)
predict=makes(model,X)
prediction_details=results(df,predict)
histogram(prediction_details)
```
